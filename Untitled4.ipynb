```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the dataset
url = 'path_to_your_downloaded_file/data.csv'  # Replace with the actual path to your file
data = pd.read_csv(url)

# Drop the last column if it's not needed for clustering
data = data.iloc[:, :-1]

# Standardize the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Set the number of clusters
k = 3  # Set the number of clusters you want
kmeans = KMeans(n_clusters=k, random_state=42)
data['Cluster'] = kmeans.fit_predict(data_scaled)

# Reduce data to 2D for visualization using PCA
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=data['Cluster'], cmap='viridis', marker='o')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title(f'K-Means Clustering with k={k}')
plt.colorbar(label='Cluster')
plt.show()


```

```python
# prompt: Implement the K-Means algorithm for clustering the dataset given at the following
# link after omitting the last attribute:
# https://drive.google.com/file/d/1O_NwpJT-8xGfU_-3llUl2sgPu0xllOrX/view?usp=sharing,GIVE CODE IN PYTHON TO FIND VALUE OF K and give graph showing clusters

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the dataset
url = '/content/USA_Housing.csv'  # Replace with your local path
data = pd.read_csv(url)

# Drop the last column
data = data.iloc[:, :-1]

# Standardize the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Determine the optimal number of clusters using the Elbow method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()

# Based on the Elbow curve, choose the optimal k
# (The point where the inertia starts to decrease less rapidly)
k = 3  # Example: Choosing k=3 based on the Elbow curve

# Apply K-Means with the chosen k
kmeans = KMeans(n_clusters=k, random_state=42)
data['Cluster'] = kmeans.fit_predict(data_scaled)

# Reduce data to 2D for visualization using PCA
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=data['Cluster'], cmap='viridis', marker='o')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title(f'K-Means Clustering with k={k}')
plt.colorbar(label='Cluster')
plt.show()
```
